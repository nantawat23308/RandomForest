```python
# import module
import numpy as np
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, PredefinedSplit, GridSearchCV
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
```

```python
air_data = pd.read_csv("Invistico_Airline.csv")
```

```python
air_data_subset = air_data.dropna(axis=0)
```

```python
air_data_subset.isna().sum()
```

```python
# encode object data to numerical (category)
air_data_subset_dummy = pd.get_dummies(air_data_subset, columns=["Customer Type", "Type of Travel", "Class"])
```

```python
air_data_subset_dummy.head()
```

```python
# True is 1, False is 0
```

```python
air_data_subset_dummy.info()
```

```python
y = air_data_subset_dummy['satisfaction']
X = air_data_subset_dummy.drop("satisfaction", axis=1)
```

```python
# split data to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# split data to validat and train from train data
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state = 0)
```

```python
# init param need to pass to grid search
cv_params = {'n_estimators' : [50,100], 
             'max_depth' : [10,50],
             'min_samples_leaf' : [0.5,1], 
             'min_samples_split' : [0.001, 0.01],
             'max_features' : ["sqrt"], 
             'max_samples' : [.5,.9]}
```

```python
# init cv for grid search parameter use PredefinedSplit
split_index = [0 if x in X_val.index else -1 for x in X_train.index]
custom_split = PredefinedSplit(split_index)
```

```python
# model
rf = RandomForestClassifier(random_state=0)
```

```python
# GridSearchCV
rf_val = GridSearchCV(estimator=rf, 
                      param_grid=cv_params, 
                      cv=custom_split, 
                      refit='f1', 
                      n_jobs = -1, 
                      verbose = 1)
```

```python
%%time
rf_val.fit(X_train, y_train)
```

```python
rf_val.best_params_
```

```python
rf_opt = RandomForestClassifier(n_estimators = 50,
                                max_depth = 50, 
                                min_samples_leaf = 1,
                                min_samples_split = 0.001,
                                max_features="sqrt",
                                max_samples = 0.9, 
                                random_state = 0)
```

```python
# after get champian model have to fit train data again
rf_opt.fit(X_train, y_train)
```

```python
y_pred = rf_opt.predict(X_test)
```

```python
pc_test = precision_score(y_test, y_pred, pos_label= "satisfied")
print("The precision score is {pc:.3f}".format(pc = pc_test))
rc_test = recall_score(y_test, y_pred, pos_label = "satisfied")
print("The recall score is {rc:.3f}".format(rc = rc_test))
ac_test = accuracy_score(y_test, y_pred)
print("The accuracy score is {ac:.3f}".format(ac = ac_test))
f1_test = f1_score(y_test, y_pred, pos_label = "satisfied")
print("The F1 score is {f1:.3f}".format(f1 = f1_test))
```
